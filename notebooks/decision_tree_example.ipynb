```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
np.random.seed(0)
x = np.linspace(0, 3.14, 100)
y = np.cos(x) + (np.random.random(100) - np.random.random(100)) * 0.1
```

```python
def dtr(x):
    if x < np.pi / 2:
        return np.cos(0 / 2) + np.cos(np.pi / 2)
    if x > np.pi / 2:
        return np.cos(np.pi) + np.cos(np.pi / 2)
```

```python
plt.scatter(x, y)
plt.plot(x, [dtr(i) for i in x])
plt.show()
```
