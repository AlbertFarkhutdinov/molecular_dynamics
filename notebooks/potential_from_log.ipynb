```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('seaborn')
```

```python
BASE_DIRECTORY = os.path.dirname(os.getcwd())
PATH_TO_DATA = os.path.join(BASE_DIRECTORY, 'data')

PATH_TO_DATA
```

```python
filename = os.path.join(
    PATH_TO_DATA,
    '2020-12-16',
    'log_2020-12-16.log',
)
```

```python
potentials = []
```

```python
with open(filename) as file:
    for line in file:
        if 'Potential energy:' in line:
            potentials.append(line[line.index('DEBUG') :-2])
```

```python
for i, potential in enumerate(potentials):
    potentials[i] = float(potential[potential.index(':') + 2: ])

potentials = np.array(potentials)
```

```python
plt.scatter(
    np.arange(potentials.size) * 0.005 + 0.005,
    potentials,
    s=5,
    color='black'
);
plt.xlabel(r'Time, $\tau$');
plt.ylabel(r'Pressure, $\epsilon / \sigma^3 $');
plt.xlim(left=0);
# plt.ylim(bottom=0, top=5);
```
