```python
dataframes = []
for filename in os.listdir():
    if filename.startswith('system_parameters'):
        if len(filename.split('_')) > 5:
            dataframes.append(pd.read_csv(filename, sep=';'))
```

```python
data = pd.concat(dataframes).reset_index(drop=True)
data.time -= 0.005
data
```

```python
data = data.loc[(data.index % 10 == 0)].reset_index(drop=True)
data.time = data.time.round(2)
data
```
