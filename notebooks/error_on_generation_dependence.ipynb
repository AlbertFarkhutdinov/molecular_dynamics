```python
import os

import matplotlib.pyplot as plt
```

```python
generations = []
errors = []

with open(os.path.join(os.getcwd(), 'fullrmc_0.log'), encoding='utf8', mode='r') as file:
    for line in file:
        if line.startswith('2021-10'):
            if 'Err:' in line:
                generations.append(float(line[line.index('Gen:')+len("Gen:"):].split()[0]))
                errors.append(float(line[line.index('Err:')+len("Err:"):].strip()))
```

```python
plt.scatter(generations, errors, s=1);
plt.xscale('log');
plt.xlabel('generation');
plt.ylabel('error');
```
